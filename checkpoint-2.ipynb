**Checking for prime number**

```python
import math

def isprime(x):
    if(x == 1 or x == 0):
        return False
    for i in  range(2, int(math.sqrt(x))+1):
        if(x%i == 0):
            return False
    return True

if(isprime(4)):
    print("Given number is prime")
else:
    print("Given number is not prime")
```

**Exercise 1. Relu Function**

```python
def relu(x):
    if(x <= 0):
        return 0
    else:
        return x
    
print(relu(-3))
```
